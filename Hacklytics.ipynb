# **Data Collection**

```python
from google.colab import drive
from bs4 import BeautifulSoup
import requests, re
import pandas as pd
import numpy as np
def html_cleaner():
    col_title = []
    countries = []
    list_of_dat1 = []
    list_of_dat2 = []
    final_dat = []
    req = requests.get("https://stats.oecd.org/index.aspx?DataSetCode=BLI#")
    soup = BeautifulSoup(req.text, "html.parser")
    my_table = soup.find("table", class_ = "DataTable")
    heading = my_table.find("thead")
    find_tr = heading.find_all("tr")
    find_th = find_tr[3].find_all("th", class_ = "HDim")
    for item in find_th:
        text = item.text
        col_title.append(text)
    body = my_table.find("tbody")
    find_tr2 = body.find_all("tr")
    for item in range(0, len(find_tr2) - 4):
        country = find_tr2[item].find("td", colspan = "2")
        countries.append(country.text)
        if item % 2 == 0:
            data = find_tr2[item].find_all("td", class_ = "Data")
            listy = []
            for i in data:
                listy.append(i.text)
            list_of_dat1.append(listy)
        else:
            data = find_tr2[item].find_all("td", class_ = "Data2")
            listy_2 = []
            for i in data:
                listy_2.append(i.text)
            list_of_dat2.append(listy_2)
    zipping = list(zip(list_of_dat1, list_of_dat2))
    for tuplea in zipping:
        for lista in tuplea:
            final_dat.append(lista)
    for a_list in final_dat:
        for i in range(len(a_list)):
            if a_list[i] == "..":
                a_list[i] = "N/A"
    df = pd.DataFrame(data = final_dat, columns = col_title, index = countries)
    df.drop(["Dwellings without basic facilities", "Housing expenditure", "Rooms per person", "Household net adjusted disposable income" \
        , "Household net wealth", "Labour market insecurity", "Employment rate", "Long-term unemployment rate", "Educational attainment" \
        , "Student skills", "Years in education", "Water quality", "Stakeholder engagement for developing regulations", "Voter turnout" \
        , "Life expectancy", "Self-reported health", "Feeling safe walking alone at night", "Homicide rate"], axis = 1, inplace = True)
    return df
    
 ########### Website Link #############   
#https://stats.oecd.org/index.aspx?DataSetCode=BLI#


############ Function Call ############
html_cleaner()
```

# **Data Manipulation**

```python
import requests, re
import pandas as pd
import numpy as np
def insight_2():
    df_1 = pd.read_csv("api_clean.csv")
    country_list_1 = []
    pollution_list = []
    countries_1 = df_1.iloc[:,0]
    for item in countries_1:
        country_list_1.append(item)
    pollution = df_1.iloc[:,3]
    for item in pollution:
        pollution_list.append(item)
    zipping_1 = list(zip(pollution_list,country_list_1))
    
    
    
    df_2 = pd.read_csv("world_happiness.csv")
    happiness_list = []
    country_list_2 = []
    happiness = df_2.iloc[:,2]
    for item in happiness:
        happiness_list.append(item)
    countries_2 = df_2.iloc[:,1]
    for item in countries_2:
        country_list_2.append(item)
    
    zipping_2 = list(zip(country_list_2,happiness_list))
    updated_list = []
    for tup in zipping_2:
        if tup[0] in country_list_1:
            updated_list.append(list(tup))
    
    for tup_1 in zipping_1:
        for a_list in updated_list:
            if tup_1[1] == a_list[0]:
                a_list.append(tup_1[0])
    
    new_df = pd.DataFrame(data = updated_list, columns = ["Country Name", "Happiness Score", "Air Pollution"])
    return new_df




############ Function Call ############
insight_2()
```

# **Data Visualization**

```python
import plotly.express as px
import pandas as pd
def visual2():
    df = pd.read_csv("api_clean.csv")
    fig = px.scatter(df, x = df.loc[:,"Air pollution"], y = df.loc[:,"Life satisfaction"], trendline = "ols", trendline_color_override="black", \
                     labels = {"Air pollution": "Air pollution", "Life satisfaction": "Life Satisfaction"}, title = "Life Satisfaction vs. Air Pollution")
    return fig.show()




############ Function Call ############
visual2()
```
